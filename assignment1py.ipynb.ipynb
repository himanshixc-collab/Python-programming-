```python
# Write a Python program that takes user input for two numbers and performs all arithmetic operations on them
#Display the results with appropriate messages for each operation
```

```python
num1=int(input("enter the first number:"))
num2=int(input("enter the second number:"))

#summation
sum=num1+num2
print("the sum is:" , sum)

#difference
diff=num1-num2
print("the difference is:", diff)

#product
prod=num1*num2
print("the product is:", prod)

#division
div=num1/num2
print("the division is:", div)

#modulus
mod=num1%num2
print("the modulus is:", mod)

#exponentination
expo=num1**num2
print("the exponent is:", expo)

#floor division
fd=num1//num2
print("the floor division is:", fd)
```

```python
# Write a python program to interchange a list items with & without using temporary values
```

```python
#original list
lst=[10,20,30,40]
print("original list:",lst)

#temporary variable
temp=lst[0]
lst[0]=lst[-1]
lst[-1]=temp
print("after swap (with temporary values):",lst)

#reset list
lst=[20,30,40,50]

#without temporary variable
lst[0],lst[-1]=lst[-1],lst[0]
print("after swap(without temporary values):",lst)
```

```python
# Create a dictionary, do all operations like addition of elements, remove elements, copy item & create nested dictionaries too
```

```python
#creating a dictionary
student={
    "name" : "tina" ,
    "age" : 17 ,
    "course" : "data science"
}
print(student)
```

```python
#adding element
student["section"]="B"
student.update({"year": 2})
print(student)
```

```python
#removing element
student.pop("year")
del student ["section"]
print(student)
```

```python
#copying dictionary
student_info= student.copy()
print(student_info)
```

```python
#nested dictionary
students = {
    "student1": {"name":"yashi", "age" : 17 , "course" : "data science" } ,
    "student2": {"name":"tanvi", "age": 18, "course" : "animation" } ,
    "student3": {"name" : "lia", "age" : 19, "course" : "data science" }
}
print(students)
```
