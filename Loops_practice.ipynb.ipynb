```python
#breaking statement
i=1
while i<10:
  print(i)
  if i==4:
    break
  i+=1
```

```python
#continue statement
i=0
while i<10:
  i+=1
  if i==5:
    continue
  print(i)
```

```python
#finding sum of first n natural numbers
n=int(input("enter the last number:"))
i=1
sum=0
while (i<= n):
  sum=sum+i
  i=i+1
print("the sum is:",sum)
```

```python
#sum of squares of n natural numbers
n=int(input("enter the last number:"))
i=1
sum=0
while (i<= n):
  sum=sum+i*i
  i=i+1
print("the sum is:",sum)
```

```python
#find sum of cubes of n natural numbers
n=int(input("enter the last number:"))
i=1
sum=0
while (i<=n):
  sum=sum+i*i*i
  i=i+1
print("the sum is:",sum)
```

```python
#factorial using loop
num=int(input("enter any number:"))
fact=1
for i in range (1,num+1):
  fact=fact*i
print("factorial of a number is",fact)
```
